```python
# 실습 1.
# tic-tac-toe 구현.
import sys
import random
from operator import eq

def setOrder():
    # 랜덤을 사용하여 선 정하기.
    a = random.randint(1, 10)
    b = random.randint(1, 10)
    # 사용자가 선이면 1, 컴퓨터가 선이면 0.
    first = -1
    while b == a:
        b = random.randint(1, 10)
    select = input("(선정하기)a와 b 중 하나를 선택하세요: ")
    while 1:
        if eq(select, 'a'):
            if a > b:
                first = 1
            else:
                first = 0
            break
        elif eq(select, 'b'):
            if a > b:
                first = 0
            else:
                first = 1
            break
        select = input("(선정하기)a와 b 중 하나를 선택하세요: ")

    sys.stdout.write("a=" + str(a) + ", b=" + str(b) + "로 ")
    if first == 0:
        print ("컴퓨터가 선 입니다.")
    else:
        print ("사용자가 선 입니다.")
    return first
    
def initBoard():
    board = []
    for i in range(3):
        row_board = []
        for j in range(3):
            row_board.append(' ')
        board.append(row_board)
    return board

def display(board):
    print("컴퓨터(O) vs 사용자(X)")
    for i in range(3):
        print("---|---|---")
        print(" " + board[i][0] + " | " + board[i][1] + " | " + board[i][2])
    print("---|---|---")
    
def getUserMove(board):
    x = 0
    y = 0
    done = 0
    while done != 1:
        print("Select (x,y) coordinates, exit : (-1, -1)")
        x = int(input("x = "))
        y = int(input("y = "))
        if x == -1 and y == -1:
            return 1
        elif x > 2 or x < -1:
            print("잘못된 위치입니다.")
        elif y > 2 or y < -1:
            print("잘못된 위치입니다.")
        else:
            if board[x][y] == ' ':
                break
            else:
                print("놓을 수 없는 곳입니다.")
    board[x][y] = 'X'
    return 0

def checkProb(board, x):
    # 각 자리에 대한 가중치 검사를 한 행을 반환.
    probMap = [0, 0, 0]
    # 이미 두어진 곳 제외.
    for i in range(3):
        if board[x][i] != ' ':
            probMap[i] = -1
        else:
            probMap[i] = calcProb(board, x, i)
    return probMap

def calcProb(board, x, y):
    # 각 자리에 대한 가중치 검사.
    cnt = 0 # 컴퓨터 돌 개수.
    unt = 0 # 사용자 돌 개수.
    bnt = 0 # 빈 공간 개수.
    # 1순위: 컴퓨터가 이기는 수. (5) cnt = 2 / bnt = 1
    # 2순위: 사용자가 이기는 수. (4) unt = 2 / bnt = 1
    # 3순위: 컴퓨터가 2개가 되는 수. (3) cnt = 1 / bnt = 2
    # 4순위: 사용자가 2개가 되는 수. (2) unt = 1 / bnt = 2
    # 5순위: 아무데나. (1)
    prob = 1 # 일단 5순위로 넣고 최선의 수를 우선순위에 넣음.
    # 가로 검사.
    for i in range(3):
        if board[x][i] == 'O':
            cnt = cnt + 1
        elif board[x][i] == 'X':
            unt = unt + 1
        else:
            bnt = bnt + 1
    tmp = retProb(cnt, unt, bnt)
    if prob < tmp:
        prob = tmp
    # 세로 검사.
    cnt = 0
    unt = 0
    bnt = 0
    for i in range(3):
        if board[i][y] == 'O':
            cnt = cnt + 1
        elif board[i][y] == 'X':
            unt = unt + 1
        else:
            bnt = bnt + 1
    tmp = retProb(cnt, unt, bnt)
    if prob < tmp:
        prob = tmp
    # 왼쪽 대각선 검사.
    cnt = 0
    unt = 0
    bnt = 0
    if x == y:
        for i in range(3):
            if board[i][i] == 'O':
                cnt = cnt + 1
            elif board[i][i] == 'X':
                unt = unt + 1
            else:
                bnt = bnt + 1
    tmp = retProb(cnt, unt, bnt)
    if prob < tmp:
        prob = tmp
    # 오른쪽 대각선 검사.
    cnt = 0
    unt = 0
    bnt = 0
    if x + y == 2:
        for i in range(3):
            if board[i][2 - i] == 'O':
                cnt = cnt + 1
            elif board[i][2 - i] == 'X':
                unt = unt + 1
            else:
                bnt = bnt + 1
    tmp = retProb(cnt, unt, bnt)
    if prob < tmp:
        prob = tmp
    return prob
    

def retProb(cnt, unt, bnt):
    # 우선순위 반환
    if cnt == 2 and bnt == 1:
        return 5
    elif unt == 2 and bnt == 1:
        return 4
    elif cnt == 1 and bnt == 2:
        return 3
    elif unt == 1 and bnt == 2:
        return 2
    else:
        return 1
    
def getComMove(board):
    probMap = []
    for i in range(3):
        probMap.append(checkProb(board, i))
    x = 0
    y = 0
    max = 0
    bnt = 0
    for i in range(3):
        for j in range(3):
            if probMap[i][j] > max:
                max = probMap[i][j]
                x = i
                y = j
            if board[i][j] == ' ':
                bnt = bnt + 1
    if bnt == 9:
        x = 1
        y = 1
    # 컴퓨터가 선일 경우 정가운데에 먼저 두도록 한다.
    # 그렇지 않으면 사용자가 이기는 경우의 수가 생긴다.
    board[x][y] = 'O'
    
def checkWin(board, player):
    # 입력 받은 player가 이겼는지 검사.
    # 결과가 나왔으면 1, 결과가 나지 않았으면 0.
    
    isWin = 0 # 이기면 1, 아니면 0.
    cnt = 0 # 컴퓨터 돌 개수.
    unt = 0 # 사용자 돌 개수.
    # 둘 중에 어떤 변수가 3이 되든간에
    # 매 턴마다 승패 여부를 검사하는 것이므로
    # 무조건 입력으로 들어온 플레이어가 승리하게 되어 있다.
    # 가로 검사.
    for i in range(3):
        if isWin == 1:
            break
        cnt = 0
        unt = 0
        for j in range(3):
            if board[i][j] == 'O':
                cnt = cnt + 1
            elif board[i][j] == 'X':
                unt = unt + 1
            if cnt == 3 or unt == 3:
                isWin = 1    
    # 세로 검사.
    for i in range(3):
        if isWin == 1:
            break
        cnt = 0
        unt = 0
        for j in range(3):
            if board[j][i] == 'O':
                cnt = cnt + 1
            elif board[j][i] == 'X':
                unt = unt + 1
            if cnt == 3 or unt == 3:
                isWin = 1
                break        
    # 좌측 대각선 검사.
    for i in range(3):
        if isWin == 1:
            break
        cnt = 0
        unt = 0
        if board[i][i] == 'O':
            cnt = cnt + 1
        elif board[i][i] == 'X':
            unt = unt + 1
        if cnt == 3 or unt == 3:
            isWin = 1
            break
    # 우측 대각선 검사.
    for i in range(3):
        if isWin == 1:
            break
        cnt = 0
        unt = 0
        if board[i][2 - i] == 'O':
            cnt = cnt + 1
        elif board[i][2 - i] == 'X':
            unt = unt + 1
        if cnt == 3 or unt == 3:
            isWin = 1
            break
    return isWin
```

```python
board = initBoard()
turn = setOrder()
isWin = 0

# 0이면 컴퓨터 차례, 1이면 사용자 차례.
for i in range(9):
    if turn == 0:
        getComMove(board)
        turn = 1
    else:
        getUserMove(board)
        turn = 0
    display(board)
    isWin = checkWin(board, turn)
    if isWin == 1:
        # (매 턴이 끝날 적마다 곧장 turn을 바꿔치기 하므로 1일 때)컴퓨터 승리.
        if turn == 1:
            print ("=====컴퓨터 승리======")
            break
        else:
            print ("=====사용자 승리======")
            break
if isWin == 0:
    print ("======비겼습니다=====")
            
```

```python
# 실습 2.
# 임의의 자연수 n이 소수인지 확인하는 함수 프로그램.

def primeNum(n):
    prime = True
    for i in range(2, n):
        if n % i == 0:
            prime = False
            break
    if prime:
        print (n, "은 소수입니다.")
    else:
        print (n, "은 소수가 아닙니다.")
```

```python
n = int(input("자연수 n을 입력해주세요: "))
primeNum(n)
```

```python
# 실습 2.
# 리스트를 정렬하는 함수 프로그램.

def sort(lst):
    for i in range(len(lst)):
        for j in range(i, len(lst)):
            if lst[i] > lst[j]:
                tmp = lst[i]
                lst[i] = lst[j]
                lst[j] = tmp
```

```python
lst = [2, 5, 3, 4, 1]
print ("정렬 전:", lst)
sort(lst)
print ("정렬 후:", lst)
```
